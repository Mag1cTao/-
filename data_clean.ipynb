```python
import numpy as np
import pandas as pd
```

```python
original_data = pd.read_excel("free_time_statistics.xlsx")
original_data.head(20)
```

```python
original_data[["星期","节次"]] = original_data["时间段"].str.split("第", expand = True)
original_data.head(20)
```

```python
original_data.drop("时间段", axis = 1, inplace = True)

    
```

```python
original_data_monday = original_data[original_data["星期"] == "周一"]
```

```python
original_data_monday.set_index("节次", inplace = True)
```

```python
original_data_monday.drop("星期", axis = 1, inplace = True)

```

```python
original_data_monday.rename(columns = {"空闲学生":"周一"})
```

```python
original_data_tuesday = original_data[original_data["星期"] == "周二"]
original_data_tuesday.set_index("节次", inplace = True)
original_data_tuesday.drop("星期", axis = 1, inplace = True)
original_data_tuesday.rename(columns = {"空闲学生":"周二"})
```

```python
original_data_wednesday = original_data[original_data["星期"] == "周三"]
original_data_wednesday.set_index("节次", inplace = True)
original_data_wednesday.drop("星期", axis = 1, inplace = True)
original_data_wednesday.rename(columns = {"空闲学生":"周三"})
```

```python
original_data_thursday = original_data[original_data["星期"] == "周四"]
original_data_thursday.set_index("节次", inplace = True)
original_data_thursday.drop("星期", axis = 1, inplace = True)
original_data_thursday.rename(columns = {"空闲学生":"周四"})
```

```python
original_data_friday = original_data[original_data["星期"] == "周五"]
original_data_friday.set_index("节次", inplace = True)
original_data_friday.drop("星期", axis = 1, inplace = True)
original_data_friday.rename(columns = {"空闲学生":"周五"})
```

```python
combined = pd.concat([original_data_monday, original_data_tuesday, original_data_wednesday, original_data_thursday,original_data_friday], axis=1)
```

```python
new_columns = ['周一', '周二', '周三', '周四', '周五']
combined.columns = new_columns
```

```python
combined.to_csv("final_free_students_data.csv")
```
